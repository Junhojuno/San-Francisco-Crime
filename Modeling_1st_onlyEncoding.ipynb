```python
df = pd.read_csv("train.csv", parse_dates=['Dates'])
# 연,월,주차,시간(1시간단위)
df['year'] = df['Dates'].map(lambda x: x.year)
df['month'] = df['Dates'].map(lambda x: x.month)
df['week'] = df['Dates'].map(lambda x: x.week)
df['hour'] = df['Dates'].map(lambda x: x.hour)
df.tail()
```

```python
train = df.copy()
```

```python
le = LabelEncoder().fit(train.Category)
```

```python
x_DayOfWeek = pd.get_dummies(train.DayOfWeek)
x_PdDistrict = pd.get_dummies(train.PdDistrict)
x_X = pd.DataFrame(train.X, columns=['lon'])
x_Y = pd.DataFrame(train.Y, columns=['lat'])
x_month = pd.get_dummies(train.month, prefix="", prefix_sep='month_')
x_hour = pd.get_dummies(train.hour, prefix="", prefix_sep='hour_')
```
